```python
from dataload import dmg777k
data = dmg777k()
```

```python
from preprocess import delete_number_literals, bin_numbers2, delete_literals,delete_all_literals

data = bin_numbers2(data,num_bins=10)
data = delete_number_literals(data)
#data = delete_all_literals(data)
```

```python
from omegaconf import DictConfig, OmegaConf
import hydra
import dataload
import embed
import evaluate
from utils.data_utils import data_to_kg, extract_ents
import preprocess
from sklearn.metrics import accuracy_score, confusion_matrix
import logging
from embed import RDF2Vec, RGCN

embedder = RGCN(data)

train_entities, test_entities, train_target, test_taget = extract_ents(data)  # extract necessary fields from data
entities = train_entities + test_entities  # combine entities for embedding
kg = data_to_kg(data)  # transfrom data into kg


embeddings, _ = embedder.fit_transform(kg, entities)
train_embeddings = embeddings[: len(train_entities)]
test_embeddings = embeddings[len(train_entities):]


```

```python
embeddings
```

```python
from evaluate import SVM
model = SVM()
model.fit(train_embeddings, train_target)
predictions = model.predict(test_embeddings)

print(f"Predicted {len(test_entities)} entities with an accuracy of "
        + f"{accuracy_score(test_taget, predictions) * 100 :.4f}%"
    )
print("Confusion Matrix :")
print(confusion_matrix(test_taget, predictions))
```
